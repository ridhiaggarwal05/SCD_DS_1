```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder = 'TASK 1'
file_name = 'API_SP.POP.TOTL_DS2_en_csv_v2_382887.csv'

file_path = os.path.join(folder, file_name)

data = pd.read_csv(file_path, skiprows=4)

print(data.columns)
```

```python
data = data.dropna(subset=['Country Name'])
population_2022 = data[['Country Name', '2022']].dropna()

top20 = population_2022.sort_values(by='2022', ascending=False).head(20)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top20['Country Name'], top20['2022'], color='skyblue')
plt.xticks(rotation=75)
plt.title('Top 20 Countries by Population (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```
